```python
import csv
import pandas as pd
import numpy as np
from sklearn import preprocessing
from scipy.cluster import hierarchy
from sklearn.model_selection import train_test_split
from sklearn import tree
from sklearn.tree import export_graphviz

import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler

import random

random.seed(10)

#Clean our data set for creating continuous numerical data
le = preprocessing.LabelEncoder()
bankData = pd.read_csv('/Users/figaro/Desktop/personal_git/dataSci450/Bank data.csv', header=0,index_col=None)
cols = bankData.columns.tolist()
cols = cols[-1:] + cols[:-1]
bankData = bankData[cols]
bankData_noPep = bankData[cols[1:10]]
    
categoricalData = ['pep', 'sex', 'region', 'married', 'car', 'save_act', 'current_act', 'mortgage']
numericalData = ['age', 'income', 'children']
for column in categoricalData:
    dataValues = le.fit(list(bankData[column].unique()))
    bankData[column] = le.transform(bankData[column])
    
for column in numericalData:
    dataValues = le.fit(list(bankData[column].unique()))
    bankData[column] = pd.to_numeric(bankData[column])
#print(len(bankData))

bank_header_list = bankData.columns.values.tolist()
bank_headers_noPep = bank_header_list[0:-1]

header_string = ""
for item in bank_header_list:
    header_string += ','+item
```

```python
classifiers = bankData['pep']
bankData_noPep = bankData.drop('pep', 1)
user_train, user_test, pep_train, pep_test = train_test_split(bankData_noPep, classifiers, test_size=.33, random_state=0)
```

```python
from IPython.display import Image
from IPython.display import display
from pydotplus import graphviz
from sklearn.externals.six import StringIO
from sklearn.metrics import accuracy_score
import io
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
sc.fit(user_train)
user_train_std = sc.transform(user_train)
clf = tree.DecisionTreeClassifier(criterion='entropy')
clf.fit(user_train, pep_train )
bank_decision_tree = clf.fit(user_train, pep_train)
pep_predict = bank_decision_tree.predict(user_test)
print(accuracy_score(pep_test, pep_predict))

try:

    import pydotplus

    dot_data = export_graphviz(
        clf, 
        out_file=None,
        # the parameters below are new in sklearn 0.18
        feature_names=('age', 'sex', 'region', 'income', 'married', 'children', 'car', 'save_act', 'current_act', 'mortgage'),  
        class_names=['YES','NO'],  
        filled=True,
        rounded=True)

    graph = pydotplus.graph_from_dot_data(dot_data)  
    display(Image(graph.create_png()))

except ImportError:
    print('pydotplus is not installed.')
       
```

```python
#Decision tree exploration
maxFit = {'f': 0, 'd': 0, 'mln': 0, 'a':0}
for i in range(5,11):
    for j in range(5,16):
        for k in ('entropy', 'gini'):
            clf = tree.DecisionTreeClassifier(criterion=k,
                                                max_features = i, 
                                                max_depth = j, 
                                                presort=True)
            bank_decision_tree = clf.fit(user_train, pep_train)
            pep_predict = bank_decision_tree.predict(user_test)
            accuracy = accuracy_score(pep_test, pep_predict)
            print('Max Features: %s, Max Depth: %s, "criterion: %s, Accuracy: %s' 
                  %(i,j,k,accuracy))
            if accuracy > maxFit['a']:
                maxFit['f'] = i
                maxFit['d'] = j
                maxFit['criterion'] = k
                maxFit['a'] = accuracy
print('\n\n Best Output')
print('Max Features: %s, Max Depth: %s, "criterion: %s, Accuracy: %s' 
        %(maxFit['f'],maxFit['d'],maxFit['criterion'],maxFit['a']))
```

```python
clf = tree.DecisionTreeClassifier(max_features = maxFit['f'], max_depth = maxFit['d'], presort=True)
clf.fit(user_train, pep_train)
features = ['age', 'sex', 'region', 'income', 'married', 'children', 'car', 'save_act', 'current_act', 'mortgage']  
try:

    import pydotplus

    dot_data = export_graphviz(
        clf, 
        out_file=None,
        # the parameters below are new in sklearn 0.18
        feature_names=features,
        filled=True,
        rounded=True)

    graph = pydotplus.graph_from_dot_data(dot_data)  
    display(Image(graph.create_png()))

except ImportError:
    print('pydotplus is not installed.')

for i in range(0,len(user_train.columns)):
     print('%s, %s' %(user_train.columns[i], clf.feature_importances_[i]))
```

```python
from sklearn import metrics
import matplotlib.pyplot as plt

predictions = clf.predict(user_test)
fpr, tpr, thresholds = metrics.roc_curve(pep_test, predictions)

plt.title('Receiver Operating Characteristic')
plt.plot(fpr, tpr, 'b',
label='AUC')
plt.legend(loc='lower right')
plt.plot([0,1],[0,1],'r--')
plt.xlim([-0.1,1.2])
plt.ylim([-0.1,1.2])
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.show()
```
